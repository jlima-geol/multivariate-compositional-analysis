Libraries

```python
import pandas as pd
import numpy as np
from skbio.stats.distance import DistanceMatrix
from skbio.stats.distance import permanova

import sys
import os

sys.path.append(os.path.abspath(".."))

from src.compositional_transforms import clr
```

Load clr data

```python
df = pd.read_csv("../data/processed/synthetic_geochemistry_clean.csv")
elements = [c for c in df.columns if c not in ["cluster_true"]]

df_clr = clr(df[elements])
```

Calculate Aitchison distance

```python
from scipy.spatial.distance import pdist, squareform

D = squareform(pdist(df_clr, metric="euclidean"))
dm = DistanceMatrix(D, ids=df.index.astype(str))
```

```python
print("IDs da Matriz:", dm.ids[:5])
print("Índices do DF:", df.index[:5].astype(str).tolist())
```

PERMANOVA

```python
results = permanova(dm, grouping=df["cluster_true"].values, permutations=999)
results
```
